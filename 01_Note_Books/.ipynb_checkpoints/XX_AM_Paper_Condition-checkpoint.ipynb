```python
import pandapower.networks as pn
import pandapower as pp
```

### Add PywerAM paths  

```python
net = pn.create_cigre_network_mv(with_der="all")
```

```python
net
```

```python
bus_name = []
for n in range(15):
    bus_name.append('Bus_'+str(n))
net.bus.name = bus_name
```

```python
net.ext_grid
```

```python
net.trafo.name = ['TR_0_1','TR_0_12']
net.trafo
```

```python
names = [x.replace(" ","_") for x in net.load.name.values]
net.load.name = names
net.load
```

```python
names = [x.replace(" ","_") for x in net.line.name.values]
names = [x.replace("-","_") for x in names]
net.line
```

```python
net.storage.name = ['ST_5','ST_10']
net.storage
```

```python
names = [x.replace(" ","_") for x in net.sgen.name.values]
net.sgen.name = names
net.sgen.replace[net.sgen.name=='WKA_7','WT_7']#='WT_7'
net.sgen
```

```python
pp.runpp(net)
```

```python
net.res_line
```
